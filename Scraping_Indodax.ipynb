```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time
```

# Scrapping Dataset

```python
all_reviews = []
batch_size = 200  # max per request
total_count = 15000
app_package = 'id.co.bitcoin'

for i in range(0, total_count, batch_size):
    print(f"Fetching {i+1} - {i+batch_size}")
    result, _ = reviews(
        app_package,
        lang='id',     # Bahasa Indonesia
        country='id',  # Indonesia
        sort=Sort.NEWEST,
        count=batch_size,
        filter_score_with=None  # Ambil semua skor
    )
    all_reviews.extend(result)
    time.sleep(1)  # delay biar gak ke-block Google

# Simpan ke DataFrame
df = pd.DataFrame(all_reviews)

# Simpan ke file CSV
df.to_csv('indodax_reviews.csv', index=False)
```
